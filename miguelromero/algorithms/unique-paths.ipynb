```python
from typing import List
# - Robot always move to bottom-right corner grid[m-1][n-1]
# - Robot can't move through a obstacle
# - Robot only move either down or right at any point it time
# - Calculate number of possible unique paths
# - Grid will be bidimensional always
# - Robot can exist in the equation xD

class Solution:
    memo = {}

    def uniquePathsWithObstacles(self, obstacleGrid: List[List[int]]) -> int:
        if obstacleGrid[0][0] == 1:
            return 0
        if len(obstacleGrid) == 1 and len(obstacleGrid[0]) == 1:
            return 1
        self.memo = {}
        return self.backtrack((0,0), obstacleGrid)

    def backtrack(self, state, grid):
        x, y = state
        backtrack_size_x = 0
        backtrack_size_y = 0
        memo_key = f'{x}_{y}'

        if memo_key in self.memo:
            return self.memo[memo_key]
        
        task = self.move_right_if_possible(state, grid)
        print('right', task, state, grid)
        if task == 'moved':
            backtrack_size_x = self.backtrack((x+1,y), grid)
            self.memo[memo_key] = backtrack_size_y + backtrack_size_x
        elif task == 'finish':
            return 1
        
        task = self.move_bottom_if_possible(state, grid)
        print('bottom', task, state, grid)
        if task == 'moved':
            backtrack_size_y = self.backtrack((x,y+1), grid)
            self.memo[memo_key] = backtrack_size_y + backtrack_size_x
        elif task == 'finish':
            return 1
        
        return backtrack_size_y + backtrack_size_x


    def move_right_if_possible(self, state, grid: List[List[int]]):
        x, y = state
        if x == len(grid[0]) -1 and y == len(grid) -1:
            return 'finish'
        next_x = x + 1
        if next_x >= len(grid[y]):
            return 'wall'
        if grid[y][next_x] == 1:
            return 'obstacle'
        return 'moved'

    def move_bottom_if_possible(self, state, grid: List[List[int]]):
        x, y = state
        if x == len(grid[0]) -1 and y == len(grid) -1:
            return 'finish'
        next_y = y + 1
        if next_y >= len(grid):
            return 'wall'
        if grid[next_y][x] == 1:
            return 'obstacle'
        return 'moved'

```
